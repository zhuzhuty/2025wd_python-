```python
import matplotlib.pyplot as plt 
```

### 折线图绘制以及样式设置

```python
'''
基础属性设置
color='red' : 折线的颜色
alpha=0.5 : 折线的透明度(0-1)
linestyle='--' : 折线的样式
linewidth=3 : 折线的宽度—粗细

线的样式
- 实线(solid)
-- 短线(dashed)
-. 短点相间线(dashdot)
： 虚点线(dotted)

折点的样式
o 实心圆
^ 实心三角
s 实心方块
* 星形
+ 加号
x x形
D 菱形
d 瘦菱形
h 六边形
H 大六边形
p 五边形
< 左三角
> 右三角
v 下三角
| 垂直线
_ 水平线
'''
x = range(1,8)
y=[17, 17, 18, 15, 11, 11, 13]
plt.plot(x,y,color='red',alpha=0.8,linestyle='-.',linewidth=3,marker='^')
plt.show()
```

### 图片大小的设置

```python
'''
figsize:指定figure的宽和高，单位为英寸；
dpi参数指定绘图对象的分辨率，即每英寸多少个像素，缺省值为80，1英寸等于2.5cm,A4纸是21*30cm的纸张
'''
import random
x = range(2,26,2)
y=[random.randint(15,30) for i in x]
plt.figure(figsize=(20,8),dpi=80) 
plt.plot(x,y)
#plt.show() # show会释放资源，所以在show之后的操作会无效
plt.savefig('test.png') # 也可以保存为svg格式的矢量图
plt.savefig('test.svg')
```

### 设置x轴和y轴的刻度，以及标题

```python
x = range(2,26,2)
y=[random.randint(15,30) for i in x]
plt.figure(figsize=(20,8),dpi=80)

# 设置x轴的刻度
# plt.xticks(x)
# plt.xticks(range(1,25))
# 设置y轴的刻度
# plt.yticks(y)
# plt.yticks(range(min(y),max(y)+1))

# 设置x轴的刻度
x_labels=['{}:00'.format(i) for i in x]
plt.xticks(x,x_labels,rotation=45) # rotation旋转刻度
# 设置y轴的刻度
y_labels=['{}℃'.format(i) for i in range(min(y),max(y)+1)]
plt.yticks(range(min(y),max(y)+1),y_labels)
plt.plot(x,y)

# 设置x轴和y轴的名称，以及标题
from matplotlib import font_manager
my_font=font_manager.FontProperties(fname='C:/Windows/Fonts/STXIHEI.TTF',size=18)
plt.xlabel('时间',fontproperties=my_font)
plt.ylabel('温度',fontproperties=my_font)
plt.title('温度变化图',fontproperties=my_font)
plt.show()
```

### 一图多线

```python
x = range(8,19)
y1=[131, 123, 122, 125, 126, 128, 129, 130, 127, 126, 125]
y2=[121, 113, 112, 115, 116, 118, 119, 120, 117, 116, 115]

plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y1,label='小明',color='red',linestyle='-.')
plt.plot(x,y2,label='小红',color='blue',linestyle='--')
plt.legend(prop=my_font) # 显示图例
```

### 散点图

```python
y = [11, 17, 16, 11, 12, 11, 12, 6, 6, 7, 8, 9, 12, 15, 14, 17,
     18, 21, 16, 17, 20, 14,15, 15, 15, 19, 21, 22, 22, 22,23]
x = range( 1, 32 )

plt.figure(figsize=(20,8),dpi=80)
plt.scatter(x,y,label='3月份')
x_labels=['3月{}日'.format(i) for i in x]
plt.xticks(x[::3],x_labels[::3],rotation=45,fontproperties=my_font)
y_labels=['{}℃'.format(i) for i in range(min(y),max(y)+1)]
plt.yticks(range(min(y),max(y)+1),y_labels,fontproperties=my_font)
plt.xlabel('日期',fontproperties=my_font)
plt.ylabel('温度',fontproperties=my_font)
plt.grid(alpha=0.5)
plt.show()
```
